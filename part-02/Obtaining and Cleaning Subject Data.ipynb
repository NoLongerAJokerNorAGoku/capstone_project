```python
def cleanstring(x):
    x = x.replace('<p>\r\n\t','')
    x = x.replace('<p>','')
    x = x.replace('</p>','')
    x = x.replace('<p>&quot;','')
    x = x.replace('&quot;','"')
    x = x.replace('&lsquo;',"'")
    x = x.replace('&rsquo;',"'")
    x = x.replace('<div>','')
    x = x.replace('&nbsp;','')
    x = x.replace('&#39;',"'")
    x = x.replace('&hellip;','...')
    x = x.replace('\r\n','')
    return x
    
SubjectDataFrame['statement'] = SubjectDataFrame['statement'].map(cleanstring)
```

```python
SubjectDataFrame.head()
```

```python
import regex as re

def cleandivs(x):
    cleaned = re.sub('<.+?>','',x)
    return cleaned

SubjectDataFrame['statement'] = SubjectDataFrame['statement'].map(cleandivs)
```

```python
def numbersnocommas(x):
    bignumbers = re.findall(',\d{3}',x)  #find a comma followed by three digits
    if len(bignumbers)>0:
        for number in bignumbers:
            nocomma = number.replace(',','')
            cleaned = x.replace(number,nocomma)
    else: cleaned = x
    return cleaned

SubjectDataFrame['statement'] = SubjectDataFrame['statement'].map(numbersnocommas)
```

```python
SubjectDataFrame['statement'] = SubjectDataFrame['statement'].map(lambda x: x.replace('&mdash;','-'))
```

```python
def furthercleaning(df):
    df['statement'] = df['statement'].map(lambda x: x.replace('&ldquo;','"'))
    df['statement'] = df['statement'].map(lambda x: x.replace('&rdquo;','"'))
    df['statement'] = df['statement'].map(lambda x: x.replace('&ndash;',"-"))
    df['statement'] = df['statement'].map(lambda x: x.replace('&amp;',''))
    df['statement'] = df['statement'].map(lambda x: x.replace('#&lrm;',''))
    df['statement'] = df['statement'].map(lambda x: x.replace('&aacute;','a'))
    df['statement'] = df['statement'].map(lambda x: x.replace('&oacute;','o'))
    df['statement'] = df['statement'].map(lambda x: x.replace('&uacute;','u'))
    df['statement'] = df['statement'].map(lambda x: x.replace('&eacute;','e'))
    df['statement'] = df['statement'].map(lambda x: x.replace('&ntilde;','n'))
    return df

SubjectDataFrame = furthercleaning(SubjectDataFrame)
```

```python
def fixingbackslash201(x):
    x = x.replace('\u201C','"')
    x = x.replace('\u201D','"')
    x = x.replace('\u2019',"'")
    return x

SubjectDataFrame['statement'] = SubjectDataFrame['statement'].map(fixingbackslash201)
```
