```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', None)
plt.style.use('fivethirtyeight')
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
import requests
import bs4
from bs4 import BeautifulSoup
```

```python
def inputtoDF(soup,df):
    dictionary = {'edition': soup.find('edition').find('edition').text,
                 'statement': soup.find('statement').text,
                 'statement_type': soup.find('statement_type').find('statement_type').text,
                 'statement_context': soup.find('statement_context').text,
                 'speaker': soup.find('speaker').find('first_name').text + ' ' + soup.find('speaker').find('last_name').text,
                 'speaker_party': soup.find('speaker').find('party').find('party').text,
                 'speaker_job': soup.find('speaker').find('current_job').text,
                 'subject': soup.find('subject').find('subject').text,
                 'ruling': soup.find('ruling').find('ruling').text,
                 'date': soup.find('statement_date').text}
    df = df.append(dictionary,ignore_index=True)
    return df
```

```python
TrialDataFrame = pd.DataFrame(columns=['edition',
                                       'statement',
                                       'statement_type',
                                       'statement_context',
                                       'speaker',
                                       'speaker_party',
                                       'speaker_job',
                                       'subject',
                                       'ruling',
                                       'date'])
```

```python
for i in range(1,100):
    try:
        URL = "http://www.politifact.com/api/v/2/statement/{}/?format=xml".format(str(i))
        response = requests.get(URL)
        soup1 = BeautifulSoup(response.text,"xml")
        TrialDataFrame = inputtoDF(soup1,TrialDataFrame)
    except: continue
```
