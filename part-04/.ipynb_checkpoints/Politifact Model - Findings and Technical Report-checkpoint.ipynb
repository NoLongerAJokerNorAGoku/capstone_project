### Setting up my classes and sub-pipelines

```python
#I can't get this to work, unfortunately! It's very frustrating.
#Instead, I've got this underneath as three separate pipelines
class TopOptionsDummified(BaseEstimator, TransformerMixin):
    """Takes in column and df, outputs dataframe with the top options dummified"""

    def __init__(self,set_feature):
        self.feature = set_feature
        self.topentries = None
        #self.column = None
        pass

    def extract_column(self, df, y=None):
        """Extracts the column for the feature specified"""
        column = df.loc[:,self.feature]
        return column

    def transform(self, df, y=None):
        """Returns the dummified feature, with less-common options set to 'OTHER'"""
        column = self.extract_column(df)
        column = column.map(lambda x: x if x in self.topentries else 'OTHER')
        dummydf = pd.get_dummies(column)
        dummydf.drop('OTHER',axis=1,inplace=True)
        return np.array(dummydf)

    def fit(self, df, y=None):
        """Within the specified feature, finds those unique entries with at least 100 
        associated observations; keeps those, turns the rest to 'Other'"""
        column = self.extract_column(df)
        countdf = pd.DataFrame(column.value_counts())
        self.topentries = list(countdf.loc[countdf[self.feature]>30].index)
        return self
```

```python
minnumber = 40
```

```python
class ExtractEdition(BaseEstimator,TransformerMixin):
    """Takes in df, outputs readability column"""

    def __init__(self):
        pass

    def transform(self, df, y=None):
        """Returns readability column"""
        column = df.loc[:,'edition']
        return column

    def fit(self, df, y=None):
        return self
    
class TopEditions(BaseEstimator, TransformerMixin):
    """Takes in column and df, outputs dataframe with the top options dummified"""

    def __init__(self):
        self.topentries = None
        pass

    def transform(self, column, y=None):
        """Returns the dummified feature, with less-common options set to 'OTHER'"""
        column = column.map(lambda x: x if x in self.topentries else 'OTHER')
        dummydf = pd.get_dummies(column)
        dummydf.drop('OTHER',axis=1,inplace=True)
        return np.array(dummydf)

    def fit(self, column, y=None):
        """Within the specified feature, finds those unique entries with at least 100 
        associated observations; keeps those, turns the rest to 'Other'"""
        countdf = pd.DataFrame(column.value_counts())
        self.topentries = list(countdf.loc[countdf['edition']>minnumber].index)
        return self

TopEditionsPipeline = Pipeline(steps = [('extractedition',ExtractEdition()),
                                        ('topeditions',TopEditions())])



class ExtractSpeaker(BaseEstimator,TransformerMixin):
    """Takes in df, outputs readability column"""

    def __init__(self):
        pass

    def transform(self, df, y=None):
        """Returns readability column"""
        column = df.loc[:,'speaker']
        return column

    def fit(self, df, y=None):
        return self
    
class TopSpeakers(BaseEstimator, TransformerMixin):
    """Takes in column and df, outputs dataframe with the top options dummified"""

    def __init__(self):
        self.topentries = None
        pass

    def transform(self, column, y=None):
        """Returns the dummified feature, with less-common options set to 'OTHER'"""
        column = column.map(lambda x: x if x in self.topentries else 'OTHER')
        dummydf = pd.get_dummies(column)
        dummydf.drop('OTHER',axis=1,inplace=True)
        return np.array(dummydf)

    def fit(self, column, y=None):
        """Within the specified feature, finds those unique entries with at least 100 
        associated observations; keeps those, turns the rest to 'Other'"""
        countdf = pd.DataFrame(column.value_counts())
        self.topentries = list(countdf.loc[countdf['speaker']>minnumber].index)
        return self

TopSpeakersPipeline = Pipeline(steps = [('extractspeaker',ExtractSpeaker()),
                                        ('topspeakers',TopSpeakers())])



class ExtractParty(BaseEstimator,TransformerMixin):
    """Takes in df, outputs readability column"""

    def __init__(self):
        pass

    def transform(self, df, y=None):
        """Returns readability column"""
        column = df.loc[:,'speaker_party']
        return column

    def fit(self, df, y=None):
        return self
    
class TopParty(BaseEstimator, TransformerMixin):
    """Takes in column and df, outputs dataframe with the top options dummified"""

    def __init__(self):
        self.topentries = None
        pass

    def transform(self, column, y=None):
        """Returns the dummified feature, with less-common options set to 'OTHER'"""
        column = column.map(lambda x: x if x in self.topentries else 'OTHER')
        dummydf = pd.get_dummies(column)
        dummydf.drop('OTHER',axis=1,inplace=True)
        return np.array(dummydf)

    def fit(self, column, y=None):
        """Within the specified feature, finds those unique entries with at least 100 
        associated observations; keeps those, turns the rest to 'Other'"""
        countdf = pd.DataFrame(column.value_counts())
        self.topentries = list(countdf.loc[countdf['speaker_party']>minnumber].index)
        return self

TopPartyPipeline = Pipeline(steps = [('extractparty',ExtractParty()),
                                     ('topparty',TopParty())])



class AttackOrNot(BaseEstimator,TransformerMixin):
    """Takes in df, outputs column indicating whether the observation was 
    an Attack (1) or a Claim(0)"""

    def __init__(self):
        pass

    def transform(self, df, y=None):
        """Returns column indicating whether the observation was an attack or not"""
        column = df.loc[:,'statement_type']
        column = column.map(lambda x: 1 if x=='Attack' else 0)
        column = np.array(column)
        return np.reshape(column,(len(column),1))

    def fit(self, df, y=None):
        return self
    
    

class AfterTrumpInauguration(BaseEstimator,TransformerMixin):
    """Takes in df, outputs column indicating whether the observation happened during
    Trump's presidency"""

    def __init__(self):
        pass

    def transform(self, df, y=None):
        """Returns column indicating whether the observation happened during Trump's presidency"""
        column = df.loc[:,'date']
        column = column.map(lambda x: 1 if x>datetime(2017,1,19) else 0)
        column = np.array(column)
        return np.reshape(column,(len(column),1))
    
    def fit(self, df, y=None):
        return self
    
    

class ExtractReadability(BaseEstimator,TransformerMixin):
    """Takes in df, outputs readability column"""

    def __init__(self):
        pass

    def transform(self, df, y=None):
        """Returns readability column"""
        column = df.loc[:,['dc_readability']]
        column = np.array(column)
        return np.reshape(column,(len(column),1))

    def fit(self, df, y=None):
        return self

ReadabilityPipeline = Pipeline(steps = [('getscores',ExtractReadability()),
                                        ('ss',StandardScaler())])



class ExtractStatements(BaseEstimator,TransformerMixin):
    """Takes in df, outputs readability column"""

    def __init__(self):
        pass

    def transform(self, df, y=None):
        """Returns readability column"""
        column = df.loc[:,'new_statement']
        return  column

    def fit(self, df, y=None):
        return self

class ToDense(BaseEstimator,TransformerMixin):
    
    def __init__(self):
        pass
    
    def transform(self, df, y=None):
        df = df.todense()
        return df
    
    def fit(self, df, y=None):
        return self
    
BagOfWordsPipeline = Pipeline(steps = [('getstatements',ExtractStatements()),
                                       ('cvec',CountVectorizer(stop_words='english')),
                                       ('tfidf',TfidfTransformer())
                                       #('dense',ToDense())
                                       #('ss',StandardScaler(with_mean=False))
                                      ])



class SubjectsOfStatement(BaseEstimator,TransformerMixin):
    """Takes in df, outputs only those columns for most extreme subjects"""

    def __init__(self):
        pass

    def transform(self, df, y=None):
        """Returns only columns for most extreme subjects"""
        df = df.loc[:,extremesubs]
        return np.array(df)
    
    def fit(self, df, y=None):
        return self
```

### Creating a function to test the performance of my model

```python
def testmodelperformance(model,model_name,testingx,testingy):
    
    predictions = model.predict(testingx)
    
    try: 
        probabilities = pd.DataFrame(model.predict_proba(testingx),columns=['predict_truth','predict_lie'])
        proba = True
    except: proba = False
        
    print 'The accuracy of this model on my test set: ', model.score(testingx,testingy)
    print '\n'
    print 'The classification report for this model on my test set:\n', classification_report(testingy,predictions)
    print '\n'
    
    if proba:
        
        fpr, tpr, thresholds = roc_curve(testingy,probabilities.predict_lie)
        area_under_roc = roc_auc_score(testingy, probabilities.predict_lie)
        plt.figure(figsize=[5,5])
        plt.plot(fpr,tpr,linewidth=3)
        plt.plot([0,1],[0,1],'k--',linewidth=3)
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title('ROC Curve for Lie Detection with {}\nArea Under Curve: {}'.format(model_name,area_under_roc))
        plt.show()

        print '\n'

        precision, recall, thresholds = precision_recall_curve(testingy,probabilities.predict_lie)
        area_under_prc = auc(recall,precision)
        plt.figure(figsize=[5,5])
        plt.plot(recall,precision,linewidth=3)
        plt.plot([0,1],[1,0],'k--',linewidth=3)
        plt.xlabel('Recall\n(true positive over all actual positives)')
        plt.ylabel('Precision\n(true positives over all predicted positives)')
        plt.title('Precision Recall Curve for Lie Detection with {}\nArea Under Curve: {}'.format(model_name,area_under_prc))
        plt.show()

        print '\n'
    
    testingxcopy = testingx.copy()
    testingxcopy['number_ruling'] = testingxcopy['ruling'].map(lambda x: 0 if x=='True' else 1 if x=='Mostly True' else 2 if x=='Half-True' else 3 if x=='Mostly False' else 4 if x=='False' else 5)
    testingxcopy['predictions'] = predictions
    testingxcopy['predictions'] = testingxcopy['predictions'].map(lambda x: 'Lie Detected' if x==1 else 'Truth Detected')

    groupeddf = testingxcopy.groupby(['number_ruling','predictions'])['predictions'].count()
    percentagepredictedlie = []
    for i in range(0,6):
        liespredicted = groupeddf.loc[(i,'Lie Detected')]
        total = groupeddf.loc[(i,'Truth Detected')] + groupeddf.loc[(i,'Lie Detected')]
        percentage = float(liespredicted)/total * 100
        percentagepredictedlie.append(int(round(percentage)))
    groupeddf.unstack().plot(kind='bar',stacked=True,color=['xkcd:tomato','xkcd:asparagus'],figsize=(10,5))
    plt.xticks([0,1,2,3,4,5],['True\n{}%\npredicted as lies'.format(percentagepredictedlie[0]),
                              'Mostly True\n{}%\npredicted as lies'.format(percentagepredictedlie[1]),
                              'Half-True\n{}%\npredicted as lies'.format(percentagepredictedlie[2]),
                              'Mostly False\n{}%\npredicted as lies'.format(percentagepredictedlie[3]),
                              'False\n{}%\npredicted as lies'.format(percentagepredictedlie[4]),
                              'Pants on Fire!\n{}%\npredicted as lies'.format(percentagepredictedlie[5])],rotation='horizontal')
    plt.xlabel('\nActual Politifact Ruling')
    plt.title('Predictions by Actual Politifact Ruling')
    plt.show()
```

### Adjusting the number of dummified speakers, editions, parties, and subjects

```python
minnumber = 60
inputmaxfeatures = None
inputngramrange = (1,1)

class ExtractEdition(BaseEstimator,TransformerMixin):
    """Takes in df, outputs readability column"""

    def __init__(self):
        pass

    def transform(self, df, y=None):
        """Returns readability column"""
        column = df.loc[:,'edition']
        return column

    def fit(self, df, y=None):
        return self
    
class TopEditions(BaseEstimator, TransformerMixin):
    """Takes in column and df, outputs dataframe with the top options dummified"""

    def __init__(self):
        self.topentries = None
        pass

    def transform(self, column, y=None):
        """Returns the dummified feature, with less-common options set to 'OTHER'"""
        column = column.map(lambda x: x if x in self.topentries else 'OTHER')
        dummydf = pd.get_dummies(column)
        if 'OTHER' in dummydf.columns:
            dummydf.drop('OTHER',axis=1,inplace=True)
        for column_name in self.topentries:
            if column_name in dummydf.columns: continue
            else: dummydf[column_name]=0
        return np.array(dummydf)

    def fit(self, column, y=None):
        """Within the specified feature, finds those unique entries with at least 100 
        associated observations; keeps those, turns the rest to 'Other'"""
        countdf = pd.DataFrame(column.value_counts())
        self.topentries = list(countdf.loc[countdf['edition']>minnumber].index)
        return self

TopEditionsPipeline = Pipeline(steps = [('extractedition',ExtractEdition()),
                                        ('topeditions',TopEditions())])



class ExtractSpeaker(BaseEstimator,TransformerMixin):
    """Takes in df, outputs readability column"""

    def __init__(self):
        pass

    def transform(self, df, y=None):
        """Returns readability column"""
        column = df.loc[:,'speaker']
        return column

    def fit(self, df, y=None):
        return self
    
class TopSpeakers(BaseEstimator, TransformerMixin):
    """Takes in column and df, outputs dataframe with the top options dummified"""

    def __init__(self):
        self.topentries = None
        pass

    def transform(self, column, y=None):
        """Returns the dummified feature, with less-common options set to 'OTHER'"""
        column = column.map(lambda x: x if x in self.topentries else 'OTHER')
        dummydf = pd.get_dummies(column)
        if 'OTHER' in dummydf.columns:
            dummydf.drop('OTHER',axis=1,inplace=True)
        for column_name in self.topentries:
            if column_name in dummydf.columns: continue
            else: dummydf[column_name]=0
        return np.array(dummydf)

    def fit(self, column, y=None):
        """Within the specified feature, finds those unique entries with at least 100 
        associated observations; keeps those, turns the rest to 'Other'"""
        countdf = pd.DataFrame(column.value_counts())
        self.topentries = list(countdf.loc[countdf['speaker']>minnumber].index)
        return self

TopSpeakersPipeline = Pipeline(steps = [('extractspeaker',ExtractSpeaker()),
                                        ('topspeakers',TopSpeakers())])



class ExtractParty(BaseEstimator,TransformerMixin):
    """Takes in df, outputs readability column"""

    def __init__(self):
        pass

    def transform(self, df, y=None):
        """Returns readability column"""
        column = df.loc[:,'speaker_party']
        return column

    def fit(self, df, y=None):
        return self
    
class TopParty(BaseEstimator, TransformerMixin):
    """Takes in column and df, outputs dataframe with the top options dummified"""

    def __init__(self):
        self.topentries = None
        pass

    def transform(self, column, y=None):
        """Returns the dummified feature, with less-common options set to 'OTHER'"""
        column = column.map(lambda x: x if x in self.topentries else 'OTHER')
        dummydf = pd.get_dummies(column)
        if 'OTHER' in dummydf.columns:
            dummydf.drop('OTHER',axis=1,inplace=True)
        for column_name in self.topentries:
            if column_name in dummydf.columns: continue
            else: dummydf[column_name]=0
        return np.array(dummydf)

    def fit(self, column, y=None):
        """Within the specified feature, finds those unique entries with at least 100 
        associated observations; keeps those, turns the rest to 'Other'"""
        countdf = pd.DataFrame(column.value_counts())
        self.topentries = list(countdf.loc[countdf['speaker_party']>minnumber].index)
        return self

TopPartyPipeline = Pipeline(steps = [('extractparty',ExtractParty()),
                                     ('topparty',TopParty())])



class AttackOrNot(BaseEstimator,TransformerMixin):
    """Takes in df, outputs column indicating whether the observation was 
    an Attack (1) or a Claim(0)"""

    def __init__(self):
        pass

    def transform(self, df, y=None):
        """Returns column indicating whether the observation was an attack or not"""
        column = df.loc[:,'statement_type']
        column = column.map(lambda x: 1 if x=='Attack' else 0)
        column = np.array(column)
        return np.reshape(column,(len(column),1))

    def fit(self, df, y=None):
        return self
    
    

class AfterTrumpInauguration(BaseEstimator,TransformerMixin):
    """Takes in df, outputs column indicating whether the observation happened during
    Trump's presidency"""

    def __init__(self):
        pass

    def transform(self, df, y=None):
        """Returns column indicating whether the observation happened during Trump's presidency"""
        column = df.loc[:,'date']
        column = column.map(lambda x: 1 if x>datetime(2017,1,19) else 0)
        column = np.array(column)
        return np.reshape(column,(len(column),1))
    
    def fit(self, df, y=None):
        return self
    
    

class ExtractReadability(BaseEstimator,TransformerMixin):
    """Takes in df, outputs readability column"""

    def __init__(self):
        pass

    def transform(self, df, y=None):
        """Returns readability column"""
        column = df.loc[:,['dc_readability']]
        column = np.array(column)
        return np.reshape(column,(len(column),1))

    def fit(self, df, y=None):
        return self

ReadabilityPipeline = Pipeline(steps = [('getscores',ExtractReadability()),
                                        ('ss',StandardScaler())])



class ExtractStatements(BaseEstimator,TransformerMixin):
    """Takes in df, outputs readability column"""

    def __init__(self):
        pass

    def transform(self, df, y=None):
        """Returns readability column"""
        column = df.loc[:,'new_statement']
        return  column

    def fit(self, df, y=None):
        return self

class ToDense(BaseEstimator,TransformerMixin):
    
    def __init__(self):
        pass
    
    def transform(self, df, y=None):
        df = df.todense()
        return df
    
    def fit(self, df, y=None):
        return self
    
BagOfWordsPipeline = Pipeline(steps = [('getstatements',ExtractStatements()),
                                       ('cvec',CountVectorizer(stop_words='english',max_features=inputmaxfeatures, ngram_range=inputngramrange)),
                                       ('tfidf',TfidfTransformer())
                                       #('dense',ToDense())
                                       #('ss',StandardScaler(with_mean=False))
                                      ])



class SubjectsOfStatement(BaseEstimator,TransformerMixin):
    """Takes in df, outputs only those columns for most extreme subjects"""

    def __init__(self):
        pass

    def transform(self, df, y=None):
        """Returns only columns for most extreme subjects"""
        df = df.loc[:,listofsubjects]
        return np.array(df)
    
    def fit(self, df, y=None):
        return self
```

### Trying lemmatization with my Count Vectorizer 

#### Lemmatizing my statements

First, I'm going to add a column to my dataframe that has my statements lemmatized.  In order to achieve this, I'm going to write a function that first splits the statement into words, then tags the word with a part-of-speech, lemmatizes the words, and then finally all of the words will be recombined into a single string.

```python
from textblob import TextBlob, Word
```

```python
def getlemmatized(x):
    
    x = x.decode('unicode_escape').encode('ascii','ignore')
    
    tb = TextBlob(x)
    words_and_pos = tb.tags
    lemmatized_statement = []
    
    for word, pos in words_and_pos:
        
        if pos[0]=='V': use_pos='v'
        elif pos[0]=='N': use_pos='n'
        elif pos[0]=='R': use_pos='r'
        elif pos[0]=='J': use_pos='a'
        else: use_pos=None
        
        lemmatized_word = word.lemmatize(use_pos)
        
        lemmatized_statement.append(lemmatized_word)
    
    lemmatized_statement = ' '.join(lemmatized_statement)
    
    return lemmatized_statement
```

```python
class EditionsClass(BaseEstimator, TransformerMixin):
    """Takes in df, outputs dataframe with the top editions options dummified"""

    def __init__(self):
        self.topentries = None
        pass
    
    def get_feature_names(self):
        return self.topentries

    def transform(self, df, y=None):
        """Returns the dummified feature, with less-common options set to 'OTHER'"""
        column = df.loc[:,'edition']
        column = column.map(lambda x: x if x in self.topentries else 'OTHER')
        dummydf = pd.get_dummies(column)
        if 'OTHER' in dummydf.columns:
            dummydf.drop('OTHER',axis=1,inplace=True)
        for column_name in self.topentries:
            if column_name in dummydf.columns: continue
            else: dummydf[column_name]=0
        self.topentries = list(dummydf.columns)
        return np.array(dummydf)

    def fit(self, df, y=None):
        """Within the specified feature, finds those unique entries with at least 100 
        associated observations; keeps those, turns the rest to 'Other'"""
        column = df.loc[:,'edition']
        countdf = pd.DataFrame(column.value_counts())
        self.topentries = list(countdf.loc[countdf['edition']>minnumber].index)
        return self

    

class SpeakerClass(BaseEstimator,TransformerMixin):
    """Takes in df, outputs df with top speakers dummified"""

    def __init__(self):
        self.topentries = None
        pass

    def get_feature_names(self):
        return self.topentries
    
    def transform(self, df, y=None):
        """Returns the dummified feature, with less-common options set to 'OTHER'"""
        column = df.loc[:,'speaker']
        column = column.map(lambda x: x if x in self.topentries else 'OTHER')
        dummydf = pd.get_dummies(column)
        if 'OTHER' in dummydf.columns:
            dummydf.drop('OTHER',axis=1,inplace=True)
        for column_name in self.topentries:
            if column_name in dummydf.columns: continue
            else: dummydf[column_name]=0
        self.topentries = list(dummydf.columns)
        return np.array(dummydf)

    def fit(self, df, y=None):
        """Within the specified feature, finds those unique entries with at least 100 
        associated observations; keeps those, turns the rest to 'Other'"""
        column = df.loc[:,'speaker']
        countdf = pd.DataFrame(column.value_counts())
        self.topentries = list(countdf.loc[countdf['speaker']>minnumber].index)
        return self



class PartyClass(BaseEstimator,TransformerMixin):     
    """Takes in df, outputs dataframe with the top class options dummified"""

    def __init__(self):
        self.topentries = None
        pass
    
    def get_feature_names(self):
        return self.topentries

    def transform(self, df, y=None):
        """Returns the dummified feature, with less-common options set to 'OTHER'"""
        column = df.loc[:,'speaker_party']
        column = column.map(lambda x: x if x in self.topentries else 'OTHER')
        dummydf = pd.get_dummies(column)
        if 'OTHER' in dummydf.columns:
            dummydf.drop('OTHER',axis=1,inplace=True)
        for column_name in self.topentries:
            if column_name in dummydf.columns: continue
            else: dummydf[column_name]=0
        self.topentries = list(dummydf.columns)
        return np.array(dummydf)

    def fit(self, column, y=None):
        """Within the specified feature, finds those unique entries with at least 100 
        associated observations; keeps those, turns the rest to 'Other'"""
        column = df.loc[:,'speaker_party']
        countdf = pd.DataFrame(column.value_counts())
        self.topentries = list(countdf.loc[countdf['speaker_party']>minnumber].index)
        return self
    


class AttackOrNot(BaseEstimator,TransformerMixin):
    """Takes in df, outputs column indicating whether the observation was 
    an Attack (1) or a Claim(0)"""

    def __init__(self):
        pass
    
    def get_feature_names(self):
        return ['Attack']

    def transform(self, df, y=None):
        """Returns column indicating whether the observation was an attack or not"""
        column = df.loc[:,'statement_type']
        column = column.map(lambda x: 1 if x=='Attack' else 0)
        column = np.array(column)
        return np.reshape(column,(len(column),1))

    def fit(self, df, y=None):
        return self
    
    

class AfterTrumpInauguration(BaseEstimator,TransformerMixin):
    """Takes in df, outputs column indicating whether the observation happened during
    Trump's presidency"""

    def __init__(self):
        pass
    
    def get_feature_names(self):
        return ['After_Inaug']

    def transform(self, df, y=None):
        """Returns column indicating whether the observation happened during Trump's presidency"""
        column = df.loc[:,'date']
        column = column.map(lambda x: 1 if x>datetime(2017,1,19) else 0)
        column = np.array(column)
        return np.reshape(column,(len(column),1))
    
    def fit(self, df, y=None):
        return self
    


class SubjectsOfStatement(BaseEstimator,TransformerMixin):
    """Takes in df, outputs only those columns for most extreme subjects"""

    def __init__(self):
        self.columns = None
        pass

    def get_feature_names(self):
        return list(self.columns)
    
    def transform(self, df, y=None):
        """Returns only columns for most extreme subjects"""
        df = df.loc[:,commonsubs]
        self.columns = df.columns
        return np.array(df)
    
    def fit(self, df, y=None):
        return self
```

### get_feature_names for Readability

```python
 class ReadabilityClass(BaseEstimator,TransformerMixin):
    """Takes in df, outputs readability column standardised"""

    def __init__(self):
        self.mean = None
        self.std = None
        pass

    def get_feature_names(self):
        return ['Readability']
    
    def transform(self, df, y=None):
        """Returns readability column"""
        column = df.loc[:,'dc_readability']
        sscolumn = (np.array(column) - self.mean)/self.std
        return np.reshape(sscolumn,(len(sscolumn),1))

    def fit(self, df, y=None):
        column = df.loc[:,['dc_readability']]
        self.mean = float(np.mean(column))
        self.std = float(np.std(column))
        return self
```

```python
class ExtractStatements(BaseEstimator,TransformerMixin):
    """Takes in df, outputs readability column"""

    def __init__(self):
        pass

    def transform(self, df, y=None):
        """Returns readability column"""
        column = df.loc[:,'new_statement']
        return  column

    def fit(self, df, y=None):
        return self

    
BagOfWordsPipeline = Pipeline(steps = [('getstatements',ExtractStatements()),
                                       ('cvec',CountVectorizer(stop_words='english',
                                                               max_features=inputmaxfeatures,
                                                               ngram_range=inputngramrange)),
                                       ('tfidf',TfidfTransformer())
                                      ])



class SubjectsOfStatement(BaseEstimator,TransformerMixin):
    """Takes in df, outputs only those columns for certain subjects"""

    def __init__(self):
        pass

    def transform(self, df, y=None):
        """Returns only columns for certain subjects"""
        df = df.loc[:,commonsubs]
        return np.array(df)
    
    def fit(self, df, y=None):
        return self

    
class ExtractLemmatizedStatements(BaseEstimator,TransformerMixin):
    """Takes in df, outputs readability column"""

    def __init__(self):
        pass

    def transform(self, df, y=None):
        """Returns readability column"""
        column = df.loc[:,'new_lemmatized']
        return  column

    def fit(self, df, y=None):
        return self

    
LemmatizedPipeline = Pipeline(steps = [('getstatements',ExtractLemmatizedStatements()),
                                       ('cvec',CountVectorizer( stop_words='english',
                                                                max_features=inputmaxfeatures,
                                                                ngram_range=inputngramrange)),
                                       ('tfidf',TfidfTransformer())
                                      ])
```

### Writing a function to get the whole list of feature names

```python
def getfeaturenames(fitted_pl):
    feature_name_list = []
    for i in [0,1,2,3,4,5,6]:
        temp_list = fitted_pl.named_steps['feats'].transformer_list[i][1].get_feature_names()
        transform_name = fitted_pl.named_steps['feats'].transformer_list[i][0]
        feat_list = [transform_name+': '+x for x in temp_list]
        feature_name_list = feature_name_list + feat_list
    feature_name_list = feature_name_list + fitted_pl.named_steps['feats'].transformer_list[7][1].named_steps['cvec'].get_feature_names()
    return feature_name_list
```

Obviously, feature importance is a little difficult to get meaningful interpretations from, because it only indicates the average total decrease in node impurity (weighted by the probability of reaching that node, which sklearn approximates with the proportion of observations reaching that node); it doesn't tell us whether that feature is positively or negatively correlated with a 'Lie', and it also can be misleading, because features that are correlated with each other may both independently be quite important, but once one of the features is used in the tree, the importance of the other feature is reduced since effectively the impurity it's capable of removing has already been removed, but it doesn't mean that by itself the feature isn't quite predictive of a lie. Hopefully this will be mitigated somewhat by the way that Extra Trees selects a random subset of features at each split (like in Random Forest). 

I would like to get a sense, for each of these, about whether they contribute more to a lie or a to a truth, so I'm going to borrow from the Logistic Regression model slightly just to get the sign of the coefficient for each feature.

### Random Forest

```python
minnumber=40

class EditionsClass(BaseEstimator, TransformerMixin):
    """Takes in df, outputs dataframe with the top editions options dummified"""

    def __init__(self):
        self.topentries = None
        pass
    
    def get_feature_names(self):
        return self.topentries

    def transform(self, df, y=None):
        """Returns the dummified feature, with less-common options set to 'OTHER'"""
        column = df.loc[:,'edition']
        column = column.map(lambda x: x if x in self.topentries else 'OTHER')
        dummydf = pd.get_dummies(column)
        if 'OTHER' in dummydf.columns:
            dummydf.drop('OTHER',axis=1,inplace=True)
        for column_name in self.topentries:
            if column_name in dummydf.columns: continue
            else: dummydf[column_name]=0
        self.topentries = list(dummydf.columns)
        return np.array(dummydf)

    def fit(self, df, y=None):
        """Within the specified feature, finds those unique entries with at least 100 
        associated observations; keeps those, turns the rest to 'Other'"""
        column = df.loc[:,'edition']
        countdf = pd.DataFrame(column.value_counts())
        self.topentries = list(countdf.loc[countdf['edition']>minnumber].index)
        return self

    

class SpeakerClass(BaseEstimator,TransformerMixin):
    """Takes in df, outputs df with top speakers dummified"""

    def __init__(self):
        self.topentries = None
        pass

    def get_feature_names(self):
        return self.topentries
    
    def transform(self, df, y=None):
        """Returns the dummified feature, with less-common options set to 'OTHER'"""
        column = df.loc[:,'speaker']
        column = column.map(lambda x: x if x in self.topentries else 'OTHER')
        dummydf = pd.get_dummies(column)
        if 'OTHER' in dummydf.columns:
            dummydf.drop('OTHER',axis=1,inplace=True)
        for column_name in self.topentries:
            if column_name in dummydf.columns: continue
            else: dummydf[column_name]=0
        self.topentries = list(dummydf.columns)
        return np.array(dummydf)

    def fit(self, df, y=None):
        """Within the specified feature, finds those unique entries with at least 100 
        associated observations; keeps those, turns the rest to 'Other'"""
        column = df.loc[:,'speaker']
        countdf = pd.DataFrame(column.value_counts())
        self.topentries = list(countdf.loc[countdf['speaker']>minnumber].index)
        return self



class PartyClass(BaseEstimator,TransformerMixin):     
    """Takes in df, outputs dataframe with the top class options dummified"""

    def __init__(self):
        self.topentries = None
        pass
    
    def get_feature_names(self):
        return self.topentries

    def transform(self, df, y=None):
        """Returns the dummified feature, with less-common options set to 'OTHER'"""
        column = df.loc[:,'speaker_party']
        column = column.map(lambda x: x if x in self.topentries else 'OTHER')
        dummydf = pd.get_dummies(column)
        if 'OTHER' in dummydf.columns:
            dummydf.drop('OTHER',axis=1,inplace=True)
        for column_name in self.topentries:
            if column_name in dummydf.columns: continue
            else: dummydf[column_name]=0
        self.topentries = list(dummydf.columns)
        return np.array(dummydf)

    def fit(self, column, y=None):
        """Within the specified feature, finds those unique entries with at least 100 
        associated observations; keeps those, turns the rest to 'Other'"""
        column = df.loc[:,'speaker_party']
        countdf = pd.DataFrame(column.value_counts())
        self.topentries = list(countdf.loc[countdf['speaker_party']>minnumber].index)
        return self
    


class AttackOrNot(BaseEstimator,TransformerMixin):
    """Takes in df, outputs column indicating whether the observation was 
    an Attack (1) or a Claim(0)"""

    def __init__(self):
        pass
    
    def get_feature_names(self):
        return ['Attack']

    def transform(self, df, y=None):
        """Returns column indicating whether the observation was an attack or not"""
        column = df.loc[:,'statement_type']
        column = column.map(lambda x: 1 if x=='Attack' else 0)
        column = np.array(column)
        return np.reshape(column,(len(column),1))

    def fit(self, df, y=None):
        return self
    
    

class AfterTrumpInauguration(BaseEstimator,TransformerMixin):
    """Takes in df, outputs column indicating whether the observation happened during
    Trump's presidency"""

    def __init__(self):
        pass
    
    def get_feature_names(self):
        return ['After_Inaug']

    def transform(self, df, y=None):
        """Returns column indicating whether the observation happened during Trump's presidency"""
        column = df.loc[:,'date']
        column = column.map(lambda x: 1 if x>datetime(2017,1,19) else 0)
        column = np.array(column)
        return np.reshape(column,(len(column),1))
    
    def fit(self, df, y=None):
        return self
    


class SubjectsOfStatement(BaseEstimator,TransformerMixin):
    """Takes in df, outputs only those columns for most extreme subjects"""

    def __init__(self):
        self.columns = None
        pass

    def get_feature_names(self):
        return list(self.columns)
    
    def transform(self, df, y=None):
        """Returns only columns for most extreme subjects"""
        df = df.loc[:,commonsubs]
        self.columns = df.columns
        return np.array(df)
    
    def fit(self, df, y=None):
        return self
```

```python
#minnumber=100


BagOfWordsPipeline = Pipeline(steps = [('getstatements',ExtractStatements()),
                                       ('cvec',CountVectorizer(stop_words='english',
                                                               max_features=5000)),
                                       ('tfidf',TfidfTransformer())
                                      ])

RFPipeline = Pipeline([
    ('feats', FeatureUnion([
        ('edition', EditionsClass()), 
        ('party', PartyClass()),
        ('speaker', SpeakerClass()),
        ('attack', AttackOrNot()),
        ('afterinaug', AfterTrumpInauguration()),
        ('readability', ReadabilityClass()),
        ('subject', SubjectsOfStatement()),
        ('bagwords', BagOfWordsPipeline)
    ])),
    ('rf', RandomForestClassifier(n_estimators=100,max_features=75,min_samples_split=8)) 
])

RFPipeline.fit(X_train,y_train)
```
